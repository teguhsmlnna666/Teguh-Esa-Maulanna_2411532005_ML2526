```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, accuracy_score

dataset_url = 'https://raw.githubusercontent.com/teguhsmlnna666/Teguh-Esa-Maulanna_2411532005_ML2526/refs/heads/main/Praktikum2/Breast_cancer_data.csv'
data = pd.read_csv(dataset_url)
data.head()
```

```python
# Bagi dataset menjadi data train dan data test
train, test = train_test_split(data, test_size=0.2, random_state=41)
X_train = train.iloc[:,:-1]
y_train = train.iloc[:,-1]
X_test = test.iloc[:,:-1]
y_test = test.iloc[:,-1]

# Buat model Gaussian Naive Bayes
gnb = GaussianNB()

# Latih model
gnb.fit(X_train, y_train)

# Lakukan prediksi
y_pred = gnb.predict(X_test)

# Evaluasi model
print("Akurasi =", accuracy_score(y_test, y_pred))
print("F1-Score =", f1_score(y_test, y_pred))
```

```python
from sklearn.metrics import (
    confusion_matrix,
    ConfusionMatrixDisplay
)
labels = [1, 0]
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp.plot();
```
